## A Smaller, Simpler Example: `DataFrame`s From Dictionaries

### Starting With A `Series` Object

```python
# this creates a Series (see Python Data Science Handbook)
# it associates US states to areas (squared Km?)
# Notice that the argument passed to the Series creation is a
# normal Python dictionary (with 5 entries (i.e., key-value pairs))
# the keys are strings (names of states) and the values are integers
# representing the areas of those states
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
area
```

### A Quick Look At `Series`

```python
# a `Series` object is `Pandas` version of a more interesting dictionary
# what can I do with it?
```

```python
# a range (alphanumeric)
area.loc['California':'Illinois']
```

```python
# a range based on position (half-closed interval: [a, b))
area.iloc[0:2]
```

```python
# a single value
area['California']
```

```python
# in-place modification as expected
area['California'] = 423960
area['California']
```

```python
# importantly, it has an `Index`
area.index
```

### Another `Series`

```python
# this Series contains the population numbers in each state.
# Notice that the index is the same as `area`
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
```

```python
pop
```

```python
pop.index
```

### A `DataFrame` From A Set Of `Series` Objects

```python
# instantiates a DataFrame with the columns, each associated with
# the two previous Series on area and population of US states
# They share an index, so we should have a DataFrame with 
# 5 rows and 2 columns
df = pd.DataFrame({'area':area, 'pop':pop})
df
```

```python
type(df)
```

```python
# each row is a Series, and they all share the same (column) index
df.loc['California']
```

```python
df.loc['California'].index
```

```python
type(df.loc['California'].index)
```

```python
df.loc['Florida']
```

```python
df.loc['Florida'].index is df.loc['California'].index
```

```python
type(df.loc['Florida'])
```

```python
# ...and each column is a Series, and they all share the same (row) index
df['area']
```

```python
df['area'].index
```

```python
df['area'].index is df['pop'].index
```

## Boolean Masks

```python
# just like `numpy`
df['pop'] >= 20000000
```

```python
df.loc[ df['pop'] >= 20000000 ]
```

```python
# I'd rather work with millions of people..
# Example of vectorisation
df.loc[:, 'pop'] = df.loc[:, 'pop'] / 1000000
df
```

```python
df.loc[ df['pop'] >= 20 ]
```

```python
# combining boolean arrays with & and | (use brackets)
df.loc[ (df['pop'] >= 20) & (df['area'] >= 500000) ]
```
